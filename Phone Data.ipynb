```python
colnames=df.select_dtypes(["object"]).columns
colnames
```

```python
from sklearn.preprocessing import OrdinalEncoder
le=OrdinalEncoder()
df[colnames]=le.fit_transform(df[colnames])
```

```python
bool_col = [col for col in df.columns if df[col].dtype == 'bool']
df[bool_col] = df[bool_col].astype(int)
```

```python
df.info()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Apply StandardScaler on Input X_train, X_test
from sklearn.preprocessing import StandardScaler
#create object of standardscaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

def mymodel(model):
    
    model.fit(X_train,Y_train)
    Y_pred=model.predict(X_test)
    
    tr=model.score(X_train,Y_train)
    te=model.score(X_test,Y_test)
    
    mse=mean_squared_error(Y_test,Y_pred)
    r2=r2_score(Y_test,Y_pred)
    
    print(f"\nTraining Score : {tr} \nTesting score :{te}")
    
    print(f"MSE : {mse}\nAccuracy_Scores_R2 {r2}")
    
    return model
```

```python
from sklearn.ensemble import GradientBoostingRegressor
mymodel(GradientBoostingRegressor(learning_rate=0.6))
```

```python
from xgboost import XGBRegressor
mymodel(XGBRegressor())
```
